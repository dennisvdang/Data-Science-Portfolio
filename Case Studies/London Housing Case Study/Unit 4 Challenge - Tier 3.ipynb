# Evaluating seasonal change in property prices

```python
# Extract the year and month from the 'Date' column and create new columns 'Year' and 'Month'
df_melted['Year'] = df_melted['Date'].dt.year
df_melted['Month'] = df_melted['Date'].dt.month

# Calculate the average property price for each month and reset the index
df_monthly = df_melted.groupby(['Year', 'Month'])['PropertyPrice'].mean().reset_index()

# Set the size of the figure
plt.figure(figsize=(10,6))

# Create a time series plot of the average property price
# Convert the 'Year' and 'Month' to a datetime format for plotting
plt.plot(pd.to_datetime(df_monthly[['Year', 'Month']].assign(day=1)), df_monthly['PropertyPrice'])

# Set the x and y-axis labels and the title of the plot
plt.xlabel('Year')
plt.ylabel('Average Property Price')
plt.title('Time Series of Average Property Price')

# Display the plot
plt.show()
```

##### Statistical test for seasonality

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Create a Series with the average property price, indexed by date
# Date is created from 'Year' and 'Month' columns of df_monthly DataFrame
s = pd.Series(df_monthly['PropertyPrice'].values, index=pd.to_datetime(df_monthly[['Year', 'Month']].assign(day=1)))

# Perform a seasonal decomposition of the time series
# The 'additive' model is used, which assumes that the components add together to make the observed data
decomposed = seasonal_decompose(s, model='additive')

# Plot the seasonal component of the decomposition
# This shows how the seasonality changes over time, with the average seasonal effect removed
decomposed.seasonal.plot()

# Display the plot
plt.show()
```

#### Determining which seasons have higher/lower property prices

```python
# Define a function to assign a season based on the month
def assign_season(month):
    if month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Autumn'
    else:
        return 'Winter'

# Create a new 'Season' column in the DataFrame using the 'assign_season' function
df_melted['Season'] = df_melted['Month'].apply(assign_season)

# Group the DataFrame by 'Borough' and 'Season' and calculate the average property price
df_borough_seasonal = df_melted.groupby(['Borough', 'Season'])['PropertyPrice'].mean().reset_index()

# Set the size of the figure
plt.figure(figsize=(14,8))

# Create a bar plot of the average property price by season for each borough
sns.barplot(x='Borough', y='PropertyPrice', hue='Season', data=df_borough_seasonal)

# Set the x and y-axis labels, and the title of the plot
plt.xlabel('Borough')
plt.ylabel('Average Property Price')
plt.title('Average Property Price by Season for Each Borough')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)

# Add a title to the legend
plt.legend(title='Season')

# Display the plot
plt.show()
```

Prices seem to be higher in Summer/Autumn than in Winter/Spring. Warrants a closer look...

```python
# Group the DataFrame by 'Year', 'Borough', and 'Season' and calculate the average property price
df_seasonal = df_melted.groupby(['Year', 'Borough', 'Season'])['PropertyPrice'].mean().reset_index()

# Group the DataFrame by 'Year' and 'Borough' and calculate the average property price
df_annual = df_melted.groupby(['Year', 'Borough'])['PropertyPrice'].mean().reset_index()

# Rename the 'PropertyPrice' column to 'AnnualAvgPrice'
df_annual.rename(columns={'PropertyPrice':'AnnualAvgPrice'}, inplace=True)

# Merge the df_annual DataFrame into the df_seasonal DataFrame 
df_seasonal = df_seasonal.merge(df_annual, on=['Year', 'Borough'], how='left')

# Normalize the seasonal average price for each borough and year by dividing by the annual average price
df_seasonal['NormalizedPrice'] = df_seasonal['PropertyPrice'] / df_seasonal['AnnualAvgPrice']

# Set the size of the figure
plt.figure(figsize=(15, 10))

# Create a box plot of the normalized property prices by season
sns.boxplot(x='Season', y='NormalizedPrice', data=df_seasonal, order=['Spring', 'Summer', 'Autumn', 'Winter'])

# Overlay a swarm plot on the box plot to show individual data points
sns.swarmplot(x='Season', y='NormalizedPrice', data=df_seasonal, size=2, color=".25", order=['Spring', 'Summer', 'Autumn', 'Winter'])

# Set the x and y-axis labels, and the title of the plot
plt.xlabel('Season')
plt.ylabel('Normalized Price')
plt.title('Normalized Property Prices by Season')

# Display the plot
plt.show()
```

```python
# Extract the normalized prices for each season
spring_prices = df_seasonal[df_seasonal['Season'] == 'Spring']['NormalizedPrice']
summer_prices = df_seasonal[df_seasonal['Season'] == 'Summer']['NormalizedPrice']
autumn_prices = df_seasonal[df_seasonal['Season'] == 'Autumn']['NormalizedPrice']
winter_prices = df_seasonal[df_seasonal['Season'] == 'Winter']['NormalizedPrice']

# Perform a one-way ANOVA test with the normalized prices for each season
# The null hypothesis is that all group means are equal
result = stats.f_oneway(spring_prices, summer_prices, autumn_prices, winter_prices)

# Print the result of the ANOVA test
print('ANOVA test result:', result)
```

ANOVA only tells us if there is a significant difference, but it doesn't tell us where that difference lies.
This is a post hoc analysis used following a one-way ANOVA to make pairwise comparisons between groups. The goal is to find out which specific groups' means (compared with each other) are different.

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Perform Tukey's HSD test
# This test is used to perform multiple pairwise comparison (post hoc comparison) between different groups
tukey = pairwise_tukeyhsd(endog=df_seasonal['NormalizedPrice'],    # Data
                          groups=df_seasonal['Season'],   # Groups
                          alpha=0.05)                     # Significance level

# Print the result of the Tukey HSD test
print(tukey)
```

Seasons from most expensive to least: Autumn > Summer > Winter > Spring

DataFrame that shows the average property price, the potential price savings, and the percent savings for buying in each season compared to the most expensive season. This can be useful for understanding the potential financial benefits of buying property in different seasons.

```python
# Calculate the average property price for each season
seasonal_avg_prices = df_seasonal.groupby('Season')['PropertyPrice'].mean()

# Identify the most expensive season
most_expensive_season = seasonal_avg_prices.idxmax()

# Calculate the potential savings for buying in each season compared to the most expensive season
price_savings = seasonal_avg_prices[most_expensive_season] - seasonal_avg_prices

# Calculate the average normalized price for each season
seasonal_avg_norm_prices = df_seasonal.groupby('Season')['NormalizedPrice'].mean()

# Calculate the percent savings for buying in each season compared to the most expensive season
percent_savings = (seasonal_avg_norm_prices[most_expensive_season] - seasonal_avg_norm_prices) / seasonal_avg_norm_prices[most_expensive_season] * 100

# Combine the results into a DataFrame
savings_df = pd.DataFrame({
    'Average Price': seasonal_avg_prices,
    'Price Savings': price_savings,
    'Percent Savings': percent_savings
})

# Display the DataFrame
savings_df
```

You can potentially save 4.45% by buying property in the Spring instead of Autumn.

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def create_price_ratio(df, borough_name):
    # Filter the DataFrame for the specified borough
    d = df[df['Borough'] == borough_name]
    
    # Calculate the average property price in 1998 and 2018
    avg_price_1998 = d[d['Date'].dt.year == 1998]['PropertyPrice'].mean()
    avg_price_2018 = d[d['Date'].dt.year == 2018]['PropertyPrice'].mean()
    
    # Calculate the ratio of the average price in 1998 to the average price in 2018
    ratio = avg_price_1998 / avg_price_2018
    
    # Return the price ratio
    return ratio

# Obtain a list of unique boroughs in the DataFrame
boroughs = df_melted['Borough'].unique()

# Use the create_price_ratio function to calculate the price ratio for each unique borough
# Add these results to a new DataFrame
price_ratio_df = pd.DataFrame({'Borough': boroughs,
                               'PriceRatio': [create_price_ratio(df_melted, borough) for borough in boroughs]})

# Calculate the percentage increase in property prices from 1998 to 2018
# This is done by subtracting the price ratio from 1 and then multiplying by 100
price_ratio_df['PercentageIncrease'] = (1 - price_ratio_df['PriceRatio']) * 100
```

```python
# Create a new figure
plt.figure(figsize=(12,8))

# Sort the DataFrame by 'PriceRatio' in ascending order
price_ratio_df = price_ratio_df.sort_values('PriceRatio', ascending=True)

# Create a bar plot of price ratio for each borough
plt.bar(price_ratio_df['Borough'], price_ratio_df['PriceRatio'])

# Add labels and title
plt.xlabel('Borough')
plt.ylabel('Price Ratio (1998 / 2018)')
plt.title('Price Ratio for Each Borough')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=90)

# Display the plot
plt.show()
```

Price ratio defined as (property price in 1998 divided by property price in 2018) can be interpreted as reverse  growth rate, with values less than 1 indicating growth. Values closer to 1 indicate a small increase in price compared to values closer to 0. Prices in Hackney increased the most while prices in North East increased the least.

```python
# Create a new figure
plt.figure(figsize=(12,8))

# Sort the DataFrame by 'PercentageIncrease' in ascending order
price_ratio_df = price_ratio_df.sort_values('PercentageIncrease', ascending=True)

# Create a bar plot of percentage increase for each borough
plt.bar(price_ratio_df['Borough'], price_ratio_df['PercentageIncrease'])

# Set the starting point of y-axis at 50%
plt.ylim([50, max(price_ratio_df['PercentageIncrease'])+10]) # Adding 10 for some margin at the top

# Add labels and title
plt.xlabel('Borough')
plt.ylabel('Percentage Increase (1998 - 2018)')
plt.title('Percentage Increase in Relative Property Price for Each Borough')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=90)

# Display the plot
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder

data = df_melted.copy()

# Convert 'Borough' from categorical to numerical using LabelEncoder
le_borough = LabelEncoder()
data['Borough'] = le_borough.fit_transform(data['Borough'])

# Define predictor variables (X) and outcome variable (y)
X = data[['Borough', 'Year', 'Month']]
y = data['PropertyPrice']

# Instantiate and fit a Random Forest model
model_rf = RandomForestRegressor(n_estimators=100, random_state=42)
model_rf.fit(X, y)

# Prepare the 2024 data. We're dropping 'PropertyPrice', 'Date', 'PostalCode', and 'Season' columns
# and setting 'Year' to 2024 for all entries
future_data_rf = data.drop(['PropertyPrice', 'Date', 'PostalCode', 'Season'], axis=1)
future_data_rf['Year'] = 2024

# Use the trained Random Forest model to predict property prices for 2024
future_prices_rf = model_rf.predict(future_data_rf)

# Add the predicted prices to the DataFrame
future_data_rf['PropertyPrice'] = future_prices_rf

# Replace the numerical 'Borough' entries with their original categorical values
future_data_rf['Borough'] = df_melted['Borough']

# Remove duplicate entries and reset index
future_data_rf = future_data_rf.drop_duplicates().reset_index(drop=True)

# Display the DataFrame
future_data_rf
```

```python
from sklearn.model_selection import cross_val_score

# Perform 10-fold cross validation on the Random Forest model
# The 'neg_mean_squared_error' scoring function is the negative of the mean squared error (MSE),
# which is a common metric for evaluating regression models
scores = cross_val_score(model_rf, X, y, cv=10, scoring='neg_mean_squared_error')

# Calculate the root mean squared error (RMSE) for each fold
# The RMSE is the square root of the MSE and provides a measure of the average error of the model's predictions
rmse_scores = np.sqrt(-scores)

# Create a box-and-whisker plot to visualize the distribution of RMSE scores
# Box plots provide a visual summary of the minimum, lower quartile, median, upper quartile, and maximum values
plt.figure(figsize=(8, 6))
plt.boxplot(rmse_scores)
plt.title('RMSE Scores from 10-Fold Cross-Validation')
plt.ylabel('RMSE')
plt.show()

# Create a histogram to visualize the distribution of RMSE scores
# Histograms divide the data into bins and show the number of observations in each bin
plt.figure(figsize=(8, 6))
plt.hist(rmse_scores, bins=10, edgecolor='black')
plt.title('Histogram of RMSE Scores from 10-Fold Cross-Validation')
plt.xlabel('RMSE')
plt.ylabel('Frequency')
plt.show()
```

Let's calculate the recovery time for each borough's property prices to return to their pre-crisis levels after a financial crisis that began in October 2008.

```python
crisis_start = pd.to_datetime('2008-10-01')
one_month_before_crisis = crisis_start - pd.DateOffset(months=1)

# Calculate the property price one month before the crisis for each borough
pre_crisis_prices = df_melted[df_melted.Date == one_month_before_crisis].set_index('Borough').PropertyPrice

# Initialize a dictionary to hold the recovery times
recovery_times = {}

# Generate feature for time it took for borough to return to pre-crisis levels
# For each borough
for borough in pre_crisis_prices.index:
    # Get the pre-crisis price
    pre_crisis_price = pre_crisis_prices.loc[borough]
    
    # Get the post-crisis prices for this borough
    post_crisis_prices = df_melted[(df_melted.Date > crisis_start) & (df_melted.Borough == borough)].sort_values('Date')
    
    # For each row in the post-crisis prices
    for i, row in post_crisis_prices.iterrows():
        # If the price is greater than or equal to the pre-crisis price
        if row.PropertyPrice >= pre_crisis_price:
            # The recovery time is the time from the crisis start to this date
            recovery_times[borough] = row.Date - crisis_start
            # Stop checking prices for this borough
            break

# Convert the recovery times dictionary to a Series
crisis_recovery_time = pd.Series(recovery_times, name='CrisisRecoveryTime')

# Convert the pre-crisis prices dictionary to a Series
pre_crisis_price = pd.Series(pre_crisis_prices, name='PreCrisisPrice')

# Merge the crisis_recovery_time Series into the original DataFrame
df_crisis = df_melted.merge(crisis_recovery_time, left_on='Borough', right_index=True, how='left')

# Merge the pre_crisis_price Series into the df_crisis DataFrame
df_crisis = df_crisis.merge(pre_crisis_price, left_on='Borough', right_index=True, how='left')
df_crisis
```

Each point on the plot represents a borough. The regression line shows the best fit quadratic relationship between the two variables.

The R-squared value is a statistical measure that represents the proportion of the variance for a dependent variable that's explained by an independent variable or variables in a regression model. If the R2 of a model is 0.50, then approximately half of the observed variation can be explained by the model's inputs.

```python
# Fit a second order polynomial to the data
coeffs = np.polyfit(grouped['PreCrisisPrice'], grouped['CrisisRecoveryTime_days'], 2)

# Calculate the predicted y-values
y_pred = np.polyval(coeffs, grouped['PreCrisisPrice'])

# Calculate the residuals
residuals = grouped['CrisisRecoveryTime_days'] - y_pred

# Calculate the R-squared
ss_res = np.sum(residuals**2)
ss_tot = np.sum((grouped['CrisisRecoveryTime_days'] - np.mean(grouped['CrisisRecoveryTime_days']))**2)
r_squared = 1 - (ss_res / ss_tot)

# Calculate the p-value
slope, intercept, r_value, p_value, std_err = stats.linregress(grouped['PreCrisisPrice'], grouped['CrisisRecoveryTime_days'])

# Create a scatter plot with a second order polynomial regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='PreCrisisPrice', y='CrisisRecoveryTime_days', data=grouped, order=2, scatter_kws={'s': 60})

# Set the title and labels
plt.title('Correlation between Pre-Crisis Borough Price and Recovery Time')
plt.xlabel('Pre-Crisis Property Price')
plt.ylabel('Recovery Time (days)')

# Add the R-squared and p-value to the plot
plt.text(0.1, 0.9, f'R-squared: {r_squared:.2f}', transform=plt.gca().transAxes)
plt.text(0.1, 0.8, f'p-value: {p_value:.2f}', transform=plt.gca().transAxes)

# Set the limits of the y-axis
plt.ylim(bottom=0)

# Show the plot
plt.show()
```

The F-statistic is a measure of statistical significance in regression models, indicating whether the data supports the model over the simpler model. The larger the F-statistic, the more likely that the variations in the data can be explained by the model.

The AIC, or Akaike information criterion, is a measure for comparing the goodness of fit of different statistical models. It takes into account both the statistical goodness of fit and the simplicity of the model. In general, a model with a lower AIC is preferred over a model with a higher AIC.

```python
import statsmodels.formula.api as smf

# Fit the first order model
model1 = smf.ols(formula='CrisisRecoveryTime_days ~ PreCrisisPrice', data=grouped).fit()

# Fit the second order model
grouped['PreCrisisPrice_squared'] = grouped['PreCrisisPrice'] ** 2
model2 = smf.ols(formula='CrisisRecoveryTime_days ~ PreCrisisPrice + PreCrisisPrice_squared', data=grouped).fit()

# Print the F-statistics and AICs
print(f"First order model F-statistic: {model1.fvalue}, AIC: {model1.aic}")
print(f"Second order model F-statistic: {model2.fvalue}, AIC: {model2.aic}") # lower AIC indicates better-fitting model
```
